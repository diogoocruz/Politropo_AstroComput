```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Abra o arquivo

arn = np.zeros(554)
with open("ThetaOrionr.txt", "r") as arquivo:

    ind =0  
    # Leia cada linha do arquivo
    for linha in arquivo:


        # Analise a linha como uma lista
        lista = eval(linha)
        print(lista[0])

        # Acesse os elementos da lista e armazene-os em variáveis individuais
        arn[ind] = lista[0]
        ind += 1
print(arn)
plt.hist(arn, bins=50)
plt.axvline(x=3.018202851614777, color="red", label="Valor de n_estrela")
plt.axvline(x=3.018202851614777+np.std(arn), color="orange", label="Desvio padrão")
plt.axvline(x=3.018202851614777-np.std(arn), color="orange")
#plt.legend()

```
